```python
import json
import os

def populate_json(domain):

    # read main task file
    main_task_file = f"{domain}.json"
    dataset_path = f"../data/{domain}/input"
    file_paths = {}
    for root, dirs, files in os.walk(dataset_path):
        for file in files:
            file_paths[file] = os.path.join(root, file).split('/input/')[1]

    if domain == 'wildfire':
        wdfr_dr = json.load(open('../dr-input/wildfire/wildfire_random.json'))

    with open(main_task_file, "r") as f:
        main_tasks = json.load(f)

    for idx, task in enumerate(main_tasks):
        task_id = task['id']
        path = f"../dr-input/{domain}/{task_id}/"
    
        if domain == 'wildfire':
            subset_files = wdfr_dr[task_id]['input_files_to_dr']
        else:
            subset_files = os.listdir(path)
            assert len(subset_files), "There is no subtask folder!"

        filepaths = []
        for f in subset_files:
            if f in file_paths.keys():
                filepaths.append(file_paths[f])
            else:
                raise Exception(f"File {f} not found in dataset files")

        main_tasks[idx]["deepresearch_subset"] = filepaths

    json.dump(main_tasks, open(main_task_file, "w"), indent=4)

# for domain in ["environment", "legal", "wildfire"]:

# legal misses subsets for question 7, 24, 25
# environment everything except 1,3,5,7

# wildfire misses downloading the relevant files from kagglehub 

for domain in ["wildfire"]:
    populate_json(domain)
```

### environment.json

```python
import json
import os
# Get list of directory names in the input directory
dr_input_dir = f"../dr-input/environment/"
all_files = os.listdir(dr_input_dir)
file_dict = {}
# Sort all_files by the last two digit of numbers separated by -
all_files.sort(key=lambda x: int(x.split('-')[-1].split('.')[0]) if '-' in x else float('inf'))
print(f"Files in {dr_input_dir}: {all_files}")
for file in all_files:
    if file.endswith('.json'): continue
    dir_path = os.path.join(dr_input_dir, file)
    if os.path.isdir(dir_path):
        print(f"Directory: {file}")
        sub_files = os.listdir(dir_path)
        # print sub_files with double quotes
        #sub_files = [f'"{f}"' for f in sub_files]
        # Add a new line after each file
        print(f"Sub-files: [{', '.join(sub_files)}]")
        file_dict[file] = sub_files
        
    else:
        print(f"{file} is not a directory.")
```

```python
file_dict["environment-easy-1"]
```

```python
json_file = f"../workload/environment.json"
with open(json_file, 'r') as f:
    data = json.load(f)

for query in data:
    queryid = query['id']
    if 'deepresearch_subset' not in query:
        query['deepresearch_subset'] = file_dict[queryid]

with open("../workload/environment_1.json", 'w') as f:
    json.dump(data, f, indent=4)
```
